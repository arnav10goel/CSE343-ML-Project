```python
import pandas as pd
import numpy as np
import torch
from transformers import AutoTokenizer, AutoModel
```

```python
train_df = pd.read_csv('train.csv')
dev_df = pd.read_csv('dev.csv')
test_df = pd.read_csv('test.csv')
```

```python
model_name = "bert-base-uncased"  # Can change to a different BERT variant
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)
```

```python
# Function to extract BERT embeddings
def get_bert_embeddings(texts):
    inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt", max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    embeddings = outputs.last_hidden_state[:, 0, :].cpu().numpy() # Extract the [CLS] embeddings
    return embeddings
```

```python
# Extract BERT embeddings for each dataset
train_embeddings = get_bert_embeddings(train_df['Text '].tolist())
dev_embeddings = get_bert_embeddings(dev_df['Text '].tolist())
test_embeddings = get_bert_embeddings(test_df['Text '].tolist())
```

```python
# Save the BERT embeddings as npy files
np.save('train_bert.npy', train_embeddings)
np.save('dev_bert.npy', dev_embeddings)
np.save('test_bert.npy', test_embeddings)
```
